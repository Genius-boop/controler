## Recréation du Projet : Analyse, Préparation des Données et Entraînement de Modèle

### 1. Chargement et Exploration Initiale des Données

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le jeu de données
data = pd.read_excel('duree de trajet.xlsx')

print('Les 5 premières lignes du jeu de données :')
display(data.head())
```

```python
print('\nForme du jeu de données :')
display(data.shape)

print('\nInformations sur le jeu de données :')
data.info()

print('\nStatistiques descriptives :')
display(data.describe())
```

### 2. Préparation des Données et Ingénierie des Caractéristiques

```python
# Sélectionner les colonnes pertinentes et gérer les valeurs manquantes
data = data[['distance_km', 'road_type', 'vehicle', 'duration_min']]
data.dropna(axis=0, inplace=True)

# Encoder les caractéristiques catégorielles : road_type et vehicle
data['road_type'] = data['road_type'].replace(['bitumée', 'non_bitumée'], [1, 0])
data['vehicle'] = data['vehicle'].replace(['moto', 'voiture'], [0, 1])

print('Les 5 premières lignes après préparation des données :')
display(data.head())
```

```python
# Séparer les caractéristiques (x) et la cible (y)
y = data['duration_min']
x = data.drop('duration_min', axis=1)

print('\nEn-tête des caractéristiques (x) :')
display(x.head())

print('\nEn-tête de la cible (y) :')
display(y.head())
```

### 3. Entraînement et Évaluation du Modèle

#### Régresseur K-plus Proches Voisins

```python
from sklearn.neighbors import KNeighborsRegressor

# Initialiser et entraîner le modèle KNeighborsRegressor
knn_model = KNeighborsRegressor()
knn_model.fit(x, y)

# Évaluer le modèle
knn_score = knn_model.score(x, y)
print(f"Score R-squared du KNeighborsRegressor : {knn_score:.4f}")
```

#### Modèle de Régression Linéaire

```python
from sklearn.linear_model import LinearRegression

# Initialiser le modèle de régression linéaire
linear_model = LinearRegression()

# Entraîner le modèle sur les données d'entraînement
linear_model.fit(x, y)

# Évaluer la performance du modèle (score R-squared)
print(f"Score R-squared du modèle de régression linéaire : {linear_model.score(x, y):.4f}")

# Afficher les coefficients du modèle
print("\nCoefficients du modèle :")
for feature, coef in zip(x.columns, linear_model.coef_):
    print(f"  {feature}: {coef:.4f}")

# Afficher l'ordonnée à l'origine du modèle
print(f"\nOrdonnée à l'origine du modèle : {linear_model.intercept_:.4f}")
```

### 4. Visualisation des Données

#### Nuage de Points : Distance vs. Durée

```python
plt.figure(figsize=(10, 6))
plt.scatter(x['distance_km'], y, alpha=0.7)
plt.title('Distance vs. Durée')
plt.xlabel('Distance (km)')
plt.ylabel('Durée (min)')
plt.grid(True)
plt.show()
```

#### Ajustement du modèle par la droite de régression

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(x['distance_km'], y, alpha=0.7, label='Données réelles')

# Prédiction avec le modèle de régression linéaire
# Nous devons créer un DataFrame pour la prédiction qui ne contient que la colonne 'distance_km'
# car le modèle a été entraîné sur toutes les caractéristiques encodées.
# Pour une visualisation simple de la relation distance-durée, nous allons simplifier ici
# en considérant les autres variables comme constantes (par exemple, road_type=1, vehicle=1).

# Pour une prédiction plus précise pour la visualisation de la relation distance-durée,
# nous allons créer des données de test avec les mêmes 'road_type' et 'vehicle' que la moyenne
# ou un scénario typique pour isoler l'effet de la distance.
# Cependant, pour ce graphe, l'objectif est de montrer l'ajustement général.

# Créons un DataFrame pour les prédictions en utilisant les valeurs de 'road_type' et 'vehicle'
# qui étaient dans les données d'entraînement pour chaque point, afin que la droite de régression
# reflète le modèle global.
# Alternativement, pour une droite de régression simple basée uniquement sur distance_km,
# nous pourrions prédire pour road_type=1 et vehicle=1.

# Pour une visualisation de la droite de régression avec la seule variable 'distance_km'
# sur le graphique 2D, il est courant de fixer les autres variables à leur moyenne ou mode,
# ou simplement de les ignorer si le modèle est majoritairement influencé par la variable visée.

# Puisque notre modèle est multi-linéaire, afficher une seule droite de régression
# pour 'distance_km' est une simplification. Pour cet exercice, nous allons
# prédire les durées en utilisant les 'road_type' et 'vehicle' des données originales
# pour voir comment le modèle global s'ajuste aux points.

y_pred_linear = linear_model.predict(x)

# Tracer la droite de régression (ou plutôt la ligne d'ajustement des prédictions)
# Ordonner les points pour que la ligne soit tracée correctement
sorted_indices = x['distance_km'].argsort()
plt.plot(x['distance_km'].iloc[sorted_indices], y_pred_linear[sorted_indices], color='red', label='Droite de Régression Linéaire')

plt.title('Ajustement du Modèle de Régression Linéaire : Distance vs. Durée')
plt.xlabel('Distance (km)')
plt.ylabel('Durée (min)')
plt.grid(True)
plt.legend()
plt.show()
```

### 5. Tester le Modèle avec de Nouvelles Données

```python
# Créer un nouveau point de données pour la prédiction
# Par exemple: distance de 50 km, route bitumée (1), véhicule voiture (1)
new_data = pd.DataFrame({
    'distance_km': [50],
    'road_type': ['bitumée'],
    'vehicle': ['voiture']
})

print('Nouvelles données brutes:')
display(new_data.head())

# Appliquer les mêmes étapes de préparation des données
# Encoder les caractéristiques catégorielles
new_data['road_type'] = new_data['road_type'].replace(['bitumée', 'non_bitumée'], [1, 0])
new_data['vehicle'] = new_data['vehicle'].replace(['moto', 'voiture'], [0, 1])

print('\nNouvelles données après préparation:')
display(new_data.head())

# Utiliser le modèle de régression linéaire pour faire une prédiction
predicted_duration = linear_model.predict(new_data)

print(f"\nDurée prédite pour les nouvelles données: {predicted_duration[0]:.2f} minutes")
```
